#  **Redes Neuronales Convolucionales (CNN)**

red-neuronal-convolucional-arquitectura.png

Una red neuronal convolucional (CNN, por sus siglas en inglés, Convolutional Neural Network) es un tipo de arquitectura de redes neuronales profundas diseñada específicamente para el procesamiento de datos con estructura de cuadrícula, como imágenes y señales de audio.

Las CNN se inspiran en la organización del sistema visual biológico y están compuestas por capas de neuronas artificiales llamadas "filtros" o "convoluciones", que se aplican a regiones solapadas de la entrada para extraer características locales. Estas características se combinan en capas posteriores para formar representaciones más abstractas y jerárquicas de los datos.

Las capas principales de una CNN son:

* Capa de convolución: Aplica filtros para detectar características específicas en la entrada.
* Capa de activación: Introduce no linealidad en la red neuronal mediante funciones de activación, como la función ReLU (Rectified Linear Unit).
* Capa de agrupación/reducción (pooling): Reduce la dimensionalidad espacial de las características y preserva la información más relevante.
* Capa totalmente conectada (Fully Connected Layer): Combina las características extraídas para realizar la clasificación o predicción final.

Las CNN han demostrado un gran rendimiento en tareas de visión por computadora, como reconocimiento de objetos, clasificación de imágenes, detección de rostros, segmentación semántica, entre otros. Además, también se utilizan en otros dominios, como el procesamiento de texto y el análisis de señales.

## Convolución
Es una operación matemática que se utiliza ampliamente en el campo del procesamiento de imágenes y en las redes neuronales convolucionales.

La convolución se realiza mediante un operador matemático conocido como kernel o filtro. El kernel es una pequeña matriz de números que se desliza (se convoluciona) a lo largo de la imagen de entrada para realizar una serie de multiplicaciones y sumas ponderadas. Estas multiplicaciones y sumas ponderadas se realizan para cada posición del kernel a lo largo de la imagen de entrada, y los resultados se acumulan para formar una nueva imagen de salida conocida como mapa de características.

El proceso de convolución permite extraer características relevantes de la imagen de entrada. El kernel actúa como un conjunto de detectores de características, y cada posición del kernel se especializa en detectar patrones o características específicas en la imagen.

## Proceso paso a paso
Se multiplica elemento a elemento las dos matrices y al final se suman los resultados y se reemplaza en la nueva matriz. En este caso se usa un kernel 3x3 y un padding de 1.

# Paso 2: Definir la arquitectura de la CNN

# ¿Cómo definimos la arquitectura de la red?


203b91f1109ebb8d98c9fa715e979336ee26821f_2_690x286.jpeg

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, 3,padding=1)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(32, 64, 3,padding=1)
        self.fc1 = nn.Linear(64 * 7 * 7, 128)
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        x = self.pool(torch.relu(self.conv1(x)))
        x = self.pool(torch.relu(self.conv2(x)))
        x = x.view(-1, 64 * 7 * 7)  # Flatten
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x

```

```python
net = Net()
print(net)
```

Se define la arquitectura de una red neuronal utilizando la clase Net (Puede utilizarse otros nombres. e.g model) que hereda de nn.Module, una clase base para definir modelos en PyTorch.

Dentro del método __init__, se definen las capas y operaciones que componen la arquitectura de la red. Cada capa se instancia como un atributo de la clase Net.

En el siguiente ejemplo, la arquitectura de la red se define de la siguiente manera:

* self.conv1 = nn.Conv2d(1, 32, 3): Se define una capa de convolución
 con 1 canal de entrada, 32 canales de salida y un kernel de tamaño 3x3. Por lo tanto, la capa nn.Conv2d(1, 32, 3) toma una imagen de entrada en escala de grises y aplica 32 filtros diferentes de tamaño 3x3 para producir 32 mapas de características diferentes. Cada mapa de características captura diferentes aspectos o patrones presentes en la imagen de entrada.

* self.pool = nn.MaxPool2d(2, 2): Se define una capa de pooling que realiza un muestreo máximo (max pooling) con un tamaño de kernel de 2x2 y un stride de 2.

Screenshot-from-2017-08-15-17-04-02.png

* self.conv2 = nn.Conv2d(32, 64, 3): Se define otra capa de convolución con 32 canales de entrada (correspondientes a los 32 canales de salida de la capa anterior), 64 canales de salida y un kernel de tamaño 3x3.

* self.fc1 = nn.Linear(64 * 7 * 7, 128): Se define una capa totalmente conectada (fully connected) que toma los datos de entrada después de las capas de convolución y pooling y los transforma en una representación de 128 dimensiones.

* self.fc2 = nn.Linear(128, 10): Se define otra capa totalmente conectada que toma la representación de 128 dimensiones y la transforma en una salida final de 10 dimensiones, que en este caso se asocia con las clases de salida de la red.

El **forward** define la secuencia de operaciones que se aplican a los datos de entrada x a medida que se propagan a través de la red neuronal.

* x = self.pool(torch.relu(self.conv1(x))): Los datos de entrada x se pasan a través de la capa de convolución self.conv1, luego se aplica la función de activación ReLU (torch.relu) a la salida de la capa de convolución y, finalmente, se aplica una capa de pooling (self.pool) para reducir las dimensiones espaciales de la salida a la mitad.

* x = self.pool(torch.relu(self.conv2(x))): La salida del paso anterior se pasa a través de otra capa de convolución self.conv2, se aplica la función de activación ReLU y, nuevamente, se aplica una capa de pooling para reducir las dimensiones espaciales de la salida a la mitad.

* x = x.view(-1, 64 * 7 * 7): La salida del paso anterior se redimensiona utilizando el método view de PyTorch. En este caso, se ajusta la forma de x a una matriz de tamaño (batch_size, 64 * 7 * 7), donde batch_size es el tamaño del lote y 64 * 7 * 7 corresponde al tamaño de las dimensiones resultantes de las capas de convolución y pooling anteriores.

* x = torch.relu(self.fc1(x)): Los datos redimensionados se pasan a través de una capa totalmente conectada (self.fc1) y se aplica la función de activación ReLU.

* x = self.fc2(x): La salida del paso anterior se pasa a través de otra capa totalmente conectada (self.fc2) sin aplicar ninguna función de activación. Esto se debe a que esta capa se utiliza para la clasificación final y, por lo general, se aplica la función de activación adecuada fuera del forward (por ejemplo, softmax) dependiendo del problema específico.

* return x: La salida final de la red se devuelve como resultado del método forward.

El **forward** también puede escribirse como una serie de paso más ordenados de la siguiente forma:

def forward(self, x):
    
  

> x = self.conv1(x)

> x = torch.relu(x)    

> x = self.pool(x)

> x = self.conv2(x)     

> x = torch.relu(x)

> x = self.pool(x)

> x = x.view(-1, 64 * 5 * 5)

> x = self.fc1(x)

> x = torch.relu(x)

> x = self.fc2(x)
  
> return x

## función de pérdida y optimizador para entrenar una red neuronal.

* criterion = nn.CrossEntropyLoss(): En esta línea, se crea una instancia de la clase nn.CrossEntropyLoss(), que es una función de pérdida comúnmente utilizada en problemas de clasificación. La función CrossEntropyLoss combina una función de activación softmax y una función de pérdida de entropía cruzada, lo que la hace adecuada para problemas en los que se busca clasificar datos en múltiples categorías.

* optimizer = optim.Adam(net.parameters(), lr=0.001): En esta línea, se define el optimizador Adam.

**net.parameters()** se utiliza para obtener los parámetros de la red que deben ser optimizados. Estos incluyen los pesos y sesgos de las capas de la red.

**lr=0.001** es el parámetro lr (tasa de aprendizaje) y determina el tamaño de los pasos que da el optimizador para actualizar los pesos de la red. Un valor más pequeño de lr implica pasos más pequeños y una convergencia más lenta, mientras que un valor más grande puede llevar a una convergencia más rápida pero también puede hacer que el entrenamiento sea inestable.

------------------------------------------------------------------

En PyTorch, existen varios criterios de pérdida (loss) y optimizadores que se pueden utilizar para entrenar redes neuronales. Algunos ejemplos disponibles en PyTorch son:

* Criterios de pérdida (Loss):

Mean Squared Error (MSE): nn.MSELoss()

Binary Cross Entropy: nn.BCELoss()

Mean Absolute Error (MAE): nn.L1Loss()

Kullback-Leibler Divergence (KL Divergence): nn.KLDivLoss()

* Optimizadores:

Gradiente Descendente Estocástico (SGD): optim.SGD()

Adam: optim.Adam()

Adagrad: optim.Adagrad()

RMSprop: optim.RMSprop()

Adadelta: optim.Adadelta()

Estos son solo algunos ejemplos de criterios de pérdida y optimizadores disponibles en PyTorch.


## Paso 4: Entrenar la red neuronal

```python
model1 = Net()

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model1.to(device)

criterion = nn.CrossEntropyLoss()  # Función de perdida
optimizer = optim.Adam(model1.parameters(), lr=0.001)  # Optimizador

n_epoch = 20   # Número de etapas de entrenamiento

for epoch in range(n_epoch):
    running_loss = 0.0
    correct = 0
    total = 0

    for i, data in enumerate(trainloader, 0):
        inputs, labels = data[0].to(device), data[1].to(device)
        optimizer.zero_grad()
        outputs = model1(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = outputs.max(1)
        total += labels.size(0)
        correct += predicted.eq(labels).sum().item()

        if i % 2000 == 1999:
            print(f"[{epoch + 1}, {i + 1}] loss: {running_loss / 2000:.3f}")
            running_loss = 0.0

    epoch_loss = running_loss / len(trainloader)
    epoch_accuracy = 100 * correct / total
    print(f"Epoch {epoch + 1}: Loss: {epoch_loss:.3f}, Accuracy: {epoch_accuracy:.2f}%")

print("Entrenamiento finalizado.")
```

```python
#@title Paso 5: Evaluar la red neuronal
model1.eval()
correct = 0
total = 0
with torch.no_grad():
    for data in testloader:
        images, labels = data[0].to(device), data[1].to(device)
        outputs = model1(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f"Precisión en el conjunto de pruebas: {100 * correct / total:.2f}%")
```

## ¿Cómo saber hasta donde entrenar?

```python
net = Net()

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
net.to(device)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(net.parameters(), lr=0.001)

n_epoch = 10
train_losses = []
test_losses = []

for epoch in range(n_epoch):
    running_loss = 0.0
    correct = 0
    total = 0

    for i, data in enumerate(trainloader, 0):
        inputs, labels = data[0].to(device), data[1].to(device)
        optimizer.zero_grad()
        outputs = net(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = outputs.max(1)
        total += labels.size(0)
        correct += predicted.eq(labels).sum().item()

        if i % 2000 == 1999:
            print(f"[{epoch + 1}, {i + 1}] loss: {running_loss / 2000:.3f}")
            running_loss = 0.0

    epoch_loss = running_loss / len(trainloader)
    epoch_accuracy = 100 * correct / total
    print(f"Epoch {epoch + 1}")
    print(f"Train Loss: {epoch_loss:.3f}, Train Accuracy: {epoch_accuracy:.2f}%")

    # Calcular la pérdida del test
    net.eval()
    test_loss = 0.0
    correct = 0
    total = 0

    with torch.no_grad():
        for data in testloader:
            images, labels = data[0].to(device), data[1].to(device)
            outputs = net(images)
            test_loss += criterion(outputs, labels).item()
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()

    test_loss /= len(testloader)
    test_accuracy = 100 * correct / total
    print(f"Test Loss: {test_loss:.3f}, Test Accuracy: {test_accuracy:.2f}%")

    # Almacenar la pérdida del entrenamiento y del test en las listas correspondientes
    train_losses.append(epoch_loss)
    test_losses.append(test_loss)

print("Entrenamiento finalizado.")

# Graficar la pérdida del entrenamiento y del test por cada epoch
plt.plot(range(1, n_epoch + 1), train_losses, label="Train Loss")
plt.plot(range(1, n_epoch + 1), test_losses, label="Test Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training and Test Loss")
plt.legend()
plt.show()
```

```python
#@title Paso 4: Entrenar la red neuronal

model2 = Net()

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model2.to(device)

criterion = nn.CrossEntropyLoss()  # Función de perdida
optimizer = optim.Adam(model2.parameters(), lr=0.001)  # Optimizador

n_epoch = 7  # Número de etapas de entrenamiento

for epoch in range(n_epoch):
    running_loss = 0.0
    correct = 0
    total = 0

    for i, data in enumerate(trainloader, 0):
        inputs, labels = data[0].to(device), data[1].to(device)
        optimizer.zero_grad()
        outputs = model2(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = outputs.max(1)
        total += labels.size(0)
        correct += predicted.eq(labels).sum().item()

        if i % 2000 == 1999:
            print(f"[{epoch + 1}, {i + 1}] loss: {running_loss / 2000:.3f}")
            running_loss = 0.0

    epoch_loss = running_loss / len(trainloader)
    epoch_accuracy = 100 * correct / total
    print(f"Epoch {epoch + 1}: Loss: {epoch_loss:.3f}, Accuracy: {epoch_accuracy:.2f}%")

print("Entrenamiento finalizado.")
```

```python
#@title Paso 5: Evaluar la red neuronal
model2.eval()
correct = 0
total = 0
with torch.no_grad():
    for data in testloader:
        images, labels = data[0].to(device), data[1].to(device)
        outputs = model2(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f"Precisión en el conjunto de pruebas: {100 * correct / total:.2f}%")
```

## ¿Cómo guardar nuestro modelo?

```python
# Guardar el modelo en un archivo
torch.save(model2.state_dict(), 'modelo.pth')
```

```python
import torch
import torch.nn as nn

# Definir la arquitectura de la red convolucional
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, 3,padding=1)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(32, 64, 3,padding=1)
        self.fc1 = nn.Linear(64 * 7 * 7, 128)
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        x = self.pool(torch.relu(self.conv1(x)))
        x = self.pool(torch.relu(self.conv2(x)))
        x = x.view(-1, 64 * 7 * 7)
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

# Crear una instancia del modelo
model_guardado = Net()

# Cargar los parámetros del modelo desde el archivo
model_guardado.load_state_dict(torch.load('modelo.pth'))
model_guardado.to(device)
```

```python
# Evaluación de la red en el conjunto de prueba
model_guardado.eval()

correct = 0
total = 0
with torch.no_grad():
    for data in testloader:
        images, labels = data[0].to(device), data[1].to(device)
        outputs = model_guardado(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f"Precisión en el conjunto de pruebas: {100 * correct / total:.2f}%")
```
