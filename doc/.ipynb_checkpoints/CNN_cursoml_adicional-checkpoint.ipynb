#  **Redes Neuronales Convolucionales (CNN)**


red-neuronal-convolucional-arquitectura.png


Una red neuronal convolucional (CNN, por sus siglas en inglés, Convolutional Neural Network) es un tipo de arquitectura de redes neuronales profundas diseñada específicamente para el procesamiento de datos con estructura de cuadrícula, como imágenes y señales de audio.

Las CNN se inspiran en la organización del sistema visual biológico y están compuestas por capas de neuronas artificiales llamadas "filtros" o "convoluciones", que se aplican a regiones solapadas de la entrada para extraer características locales. Estas características se combinan en capas posteriores para formar representaciones más abstractas y jerárquicas de los datos.

Las capas principales de una CNN son:



*   Capa de convolución: Aplica filtros para detectar características
 específicas en la entrada.
*   Capa de activación: Introduce no linealidad en la red neuronal mediante funciones de activación, como la función ReLU (Rectified Linear Unit).
*   Capa de agrupación/reducción (pooling): Reduce la dimensionalidad espacial de las características y preserva la información más relevante.
*   Capa totalmente conectada (Fully Connected Layer): Combina las características extraídas para realizar la clasificación o predicción final.

Las CNN han demostrado un gran rendimiento en tareas de visión por computadora, como reconocimiento de objetos, clasificación de imágenes, detección de rostros, segmentación semántica, entre otros. Además, también se utilizan en otros dominios, como el procesamiento de texto y el análisis de señales.




# Módulos necesarios

```python
import torch
import torchvision
import torchvision.transforms as transforms
import torch.nn as nn
import torch.optim as optim
```

En primera instancia, es necesario definir un preprocesamiento de los datos, ya que para este ejemplo trabajaremos con imágenes.

El siguiente código define una secuencia de transformaciones de datos para el preprocesamiento de imágenes. Estas transformaciones se aplicarán a las imágenes antes de ser utilizadas en un modelo de aprendizaje automático.

La primera transformación, **transforms.ToTensor()**, convierte las imágenes en tensores. En el contexto del aprendizaje automático, los tensores son estructuras de datos que representan matrices o vectores multidimensionales y son la forma de representación de datos utilizada por muchos marcos de trabajo de aprendizaje automático.

La segunda transformación, **transforms.Normalize()**, normaliza los valores de los píxeles de las imágenes. La normalización es un paso común en el preprocesamiento de imágenes, que ajusta el rango de valores de los píxeles para que estén en una escala específica. En este caso, se normaliza utilizando una media de (0.5, 0.5, 0.5) y una desviación estándar de (0.5, 0.5, 0.5).

```python
# Transformaciones de datos para preprocesamiento
transform = transforms.Compose([
    transforms.ToTensor(),  # Convertir las imágenes a tensores
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))  # Normalizar los valores de los píxeles
])
```

## **Descarga del dataset de trabajo**
Este código descarga y carga el conjunto de datos CIFAR-10 utilizando la biblioteca torchvision. CIFAR-10 es un conjunto de datos ampliamente utilizado en el campo del aprendizaje automático y la visión por computadora, que consta de 60,000 imágenes en color de 32x32 píxeles.

El conjunto de datos CIFAR-10 consta de 10 clases diferentes. Cada imagen en el conjunto de datos pertenece a una de las siguientes clases:

*   Avión (airplane)
*   Automóvil (automobile)
*   Pájaro (bird)
*   Gato (cat)
*   Ciervo (deer)
*   Perro (dog)
*   Rana (frog)
*   Caballo (horse)
*   Barco (ship)
*   Camión (truck)

Estas clases representan diferentes objetos o categorías que se encuentran comúnmente en imágenes del mundo real. Cada imagen en el conjunto de datos está etiquetada con la clase a la que pertenece, lo que permite su uso en tareas de clasificación y reconocimiento de objetos.

```python
# Descargar y cargar el conjunto de datos CIFAR-10
train_dataset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
test_dataset = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)
```

## Visualización de las imágenes

```python
import matplotlib.pyplot as plt
import numpy as np

# Crear el gráfico de 2x2 para visualizar las imágenes
fig, axs = plt.subplots(2, 2, figsize=(8, 8))

# Recorrer las primeras cuatro imágenes y sus etiquetas
for i in range(4):
    image, label = train_dataset[i]
    image = np.transpose(image, (1, 2, 0))  # Transponer los ejes de la imagen
    row = i // 2
    col = i % 2

    # Mostrar la imagen en el subplot correspondiente
    axs[row, col].imshow(image)
    axs[row, col].axis('off')
    axs[row, col].set_title(f"Label: {label}")

# Ajustar el espaciado entre los subplots y mostrar el gráfico
plt.tight_layout()
plt.show()
```

El siguiente código crea los "data loaders" (cargadores de datos) que facilitan el acceso a los datos durante el entrenamiento de un modelo de aprendizaje automático. Los data loaders se utilizan para cargar los datos en lotes (batches) de un tamaño específico y, opcionalmente, mezclarlos aleatoriamente antes de cada época de entrenamiento.



```python
# Crear los data loaders para facilitar el acceso a los datos durante el entrenamiento
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=32, shuffle=False)
```

## Definiendo la CNN

La arquitectura de la red definida en este código consta de los siguientes componentes:

* self.conv1: Una capa convolucional que realiza convoluciones en las imágenes de entrada. La capa nn.Conv2d especifica que se espera una entrada de 3 canales de color (RGB) y produce 16 mapas de características (también llamados canales) como salida. Utiliza un tamaño de kernel de 3x3 y un relleno (padding) de 1 para mantener el tamaño de la entrada.

El padding (o relleno) en el contexto de las redes neuronales convolucionales se refiere a la técnica de agregar píxeles adicionales alrededor del borde de una imagen de entrada antes de aplicar la operación de convolución. Estos píxeles adicionales suelen tener un valor de cero. El objetivo principal del padding es preservar el tamaño espacial de la imagen a medida que se aplica la convolución. Sin el padding, la aplicación de capas convolucionales reduciría progresivamente el tamaño de la imagen a medida que se desplaza el filtro sobre ella, lo que podría resultar en la pérdida de información en los bordes.
El tamaño del padding se puede ajustar y se especifica como un parámetro en la capa convolucional. Por ejemplo, en el código mencionado anteriormente, padding=1 indica que se agrega un relleno de un píxel alrededor de la imagen en cada dimensión (alto y ancho) antes de aplicar la convolución. Esto asegura que el tamaño espacial de la imagen se mantenga después de la convolución.

* self.relu: Una función de activación ReLU (Rectified Linear Unit), que aplica una función no lineal a la salida de la capa convolucional para introducir no linealidades en la red y mejorar su capacidad de aprendizaje.

* self.fc: Una capa totalmente conectada (fully connected) que realiza una transformación lineal de la salida de la capa convolucional a las clases de salida. La capa nn.Linear especifica que se espera una entrada de tamaño 16 * 32 * 32 (correspondiente al tamaño de los mapas de características) y produce una salida de tamaño num_classes, que es el número de clases en el problema de clasificación.

La función forward define la secuencia de operaciones que se realizarán cuando se pase una entrada a través de la red. En este caso, la entrada x se pasa a través de la capa convolucional, se aplica la función de activación ReLU, se aplana en un tensor unidimensional, y luego se pasa a través de la capa totalmente conectada para obtener las salidas de clasificación.

```python
# Definir la arquitectura de la red convolucional
class CNN(nn.Module):
    def __init__(self, num_classes):
        super(CNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 16, 3, padding=1)
        self.relu = nn.ReLU()
        self.fc = nn.Linear(16 * 32 * 32, num_classes)

    def forward(self, x):
        x = self.conv1(x)
        x = self.relu(x)
        x = x.view(x.size(0), -1)
        x = self.fc(x)
        return x
```

```python
# Crear una instancia de la red
num_classes = 10
model = CNN(num_classes)
```

## Optimizador
Se crea un optimizador utilizando el algoritmo de descenso de gradiente estocástico (SGD, por sus siglas en inglés) con momentum.

Los parámetros utilizados son:

* model.parameters(): Se refiere a los parámetros del modelo que se desea optimizar. Estos parámetros son los que se actualizarán durante el entrenamiento para minimizar la función de pérdida. El optimizador ajustará estos parámetros iterativamente para encontrar valores que minimicen la pérdida.

* lr=0.001: Es el valor de la tasa de aprendizaje (learning rate) que determina el tamaño de los pasos que da el optimizador para actualizar los parámetros. Controla la rapidez con la que el modelo converge hacia una solución óptima. Un valor más alto acelera el aprendizaje, pero puede hacer que el entrenamiento sea inestable, mientras que un valor más bajo hace que el entrenamiento sea más lento pero más estable.

* momentum=0.9: Es un parámetro que controla la acumulación de momento en el algoritmo de SGD con momentum. El momentum ayuda a acelerar el entrenamiento y a evitar que el optimizador quede atrapado en mínimos locales. Un valor de momentum de 0.9 significa que se tiene en cuenta el 90% del momento acumulado anteriormente y el 10% del gradiente instantáneo en cada paso.

```python
# Crear el optimizador después de definir los parámetros del modelo
optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.9)
```

Además del optimizador SGD con momentum, existen muchos otros optimizadores populares que se utilizan en el entrenamiento de modelos de aprendizaje automático. Algunos de ellos son:

* Adam (Adaptive Moment Estimation): Combina las ventajas del algoritmo AdaGrad y RMSProp. Es ampliamente utilizado y suele ser una buena elección por defecto.

* RMSProp (Root Mean Square Propagation): Adapta la tasa de aprendizaje para cada parámetro individualmente según la magnitud de sus gradientes recientes.

* Adagrad (Adaptive Gradient): Adapta la tasa de aprendizaje de forma individual para cada parámetro en función de las sumas acumulativas de los gradientes cuadráticos anteriores.

* Adadelta: Mejora el método Adagrad al resolver el problema de la reducción de la tasa de aprendizaje de forma exponencial.

* AdamW: Es similar a Adam, pero con una corrección para el sesgo de los momentos de primer y segundo orden.

* Adamax: Variante de Adam que utiliza el infinito norma en lugar de la norma 2 de los gradientes.



## Entrenamiento de la Red

Significado de las diferentes partes del código:

1.  num_epochs: Especifica el número total de épocas, es decir, la
 cantidad de veces que el modelo recorrerá todo el conjunto de datos de entrenamiento durante el entrenamiento.

2.  device = torch.device("cuda" if torch.cuda.is_available() else "cpu"): Verifica si hay una GPU disponible y, de ser así, establece el dispositivo en "cuda" para aprovechar el entrenamiento acelerado por GPU. Si no hay GPU disponible, se utiliza "cpu".

3.  model.to(device): Mueve el modelo a la GPU (si está disponible) o a la CPU para realizar los cálculos.

4.  El bucle principal for epoch in range(num_epochs): ejecuta el entrenamiento durante el número especificado de épocas.

5.  train_loss = 0.0, correct = 0, total = 0: Inicializa las variables que se utilizarán para calcular la pérdida de entrenamiento, la precisión de entrenamiento y el número total de ejemplos de entrenamiento.

6.  for images, labels in train_loader: Itera sobre el conjunto de datos de entrenamiento cargado en el train_loader. En cada iteración, se obtienen lotes de imágenes y etiquetas para el entrenamiento.

7.  images, labels = images.to(device), labels.to(device): Mueve los datos de entrada (imágenes y etiquetas) al dispositivo especificado (GPU o CPU).

8.  outputs = model(images): Pasa las imágenes a través del modelo para obtener las salidas predichas.

9.  loss = nn.CrossEntropyLoss()(outputs, labels): Calcula la pérdida utilizando la función de pérdida de entropía cruzada. Compara las salidas predichas con las etiquetas reales.

10. optimizer.zero_grad(): Limpia los gradientes acumulados de los parámetros del modelo antes de realizar la retropropagación.

11. loss.backward(): Realiza la retropropagación para calcular los gradientes de los parámetros del modelo en función de la pérdida. La función loss.backward() se utiliza durante el entrenamiento de una red neuronal para calcular cómo los cambios en los pesos y sesgos de la red afectan a la pérdida o error en la salida. La retropropagación es el proceso mediante el cual se calculan estos gradientes de los parámetros del modelo en función de la pérdida. Durante la retropropagación, se utiliza la función de pérdida para medir el error entre las salidas predichas por la red y las etiquetas reales de los datos de entrenamiento. Luego, se calculan los gradientes que indican cómo los parámetros de la red deben ajustarse para reducir la pérdida.

12. optimizer.step(): Actualiza los parámetros del modelo utilizando el optimizador.

13. _, predicted = torch.max(outputs.data, 1): Calcula las etiquetas predichas tomando el índice de la clase con el valor máximo de las salidas del modelo.

14. total += labels.size(0), correct += (predicted == labels).sum().item(): Actualiza las variables total y correct para calcular la precisión de entrenamiento.

15. train_loss += loss.item(): Acumula la pérdida de entrenamiento para calcular la pérdida promedio después de cada época.

Al final de cada época, se imprime la pérdida promedio de entrenamiento y la precisión de entrenamiento.



```python
# Entrenamiento de la red
num_epochs = 5
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

train_losses = []
train_accuracies = []

for epoch in range(num_epochs):
    train_loss = 0.0
    correct = 0
    total = 0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        # Calcular la salida y la pérdida
        outputs = model(images)
        loss = nn.CrossEntropyLoss()(outputs, labels)

        # Realizar la retropropagación y la optimización
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Calcular la precisión del entrenamiento
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

        train_loss += loss.item()

    # Calcular la precisión y pérdida promedio en el conjunto de entrenamiento para cada época
    train_losses.append(train_loss / len(train_loader))
    train_accuracies.append((correct / total) * 100)

    # Imprimir métricas del entrenamiento
    print(f"Epoch [{epoch+1}/{num_epochs}], Train Loss: {train_loss/len(train_loader):.4f}, Train Accuracy: {(correct/total)*100:.2f}%")
```

```python
import matplotlib.pyplot as plt
# Graficar la pérdida y precisión en el conjunto de entrenamiento a lo largo de las épocas
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(train_losses)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Train Loss')

plt.subplot(1, 2, 2)
plt.plot(train_accuracies)
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Train Accuracy')

plt.tight_layout()
plt.show()
```

## Evaluación (Testing)
Significado de las diferentes partes del código:

1. model.eval(): Establece el modelo en modo de evaluación. Esto es importante porque algunos capas, como la capa de dropout o la normalización por lotes, pueden tener un comportamiento diferente durante la evaluación en comparación con el entrenamiento.

2. test_loss = 0.0, correct = 0, total = 0: Inicializa las variables que se utilizarán para calcular la pérdida en el conjunto de prueba, la precisión en el conjunto de prueba y el número total de ejemplos en el conjunto de prueba.

3. with torch.no_grad(): Este contexto desactiva el seguimiento de gradientes en el bloque de código que lo rodea. Esto es útil durante la evaluación, ya que no se necesitan los gradientes y desactivar el seguimiento de gradientes puede ahorrar memoria y acelerar el proceso.

4. for images, labels in test_loader: Itera sobre el conjunto de datos de prueba cargado en el test_loader. En cada iteración, se obtienen lotes de imágenes y etiquetas para la evaluación.

5. images, labels = images.to(device), labels.to(device): Mueve los datos de entrada (imágenes y etiquetas) al dispositivo especificado (GPU o CPU).

6. outputs = model(images): Pasa las imágenes a través del modelo para obtener las salidas predichas.

7. loss = nn.CrossEntropyLoss()(outputs, labels): Calcula la pérdida en función de las salidas predichas y las etiquetas reales utilizando la función de pérdida de entropía cruzada.

8. _, predicted = torch.max(outputs.data, 1): Calcula las etiquetas predichas tomando el índice de la clase con el valor máximo de las salidas del modelo.

9. total += labels.size(0), correct += (predicted == labels).sum().item(): Actualiza las variables total y correct para calcular la precisión en el conjunto de prueba.

10. test_loss += loss.item(): Acumula la pérdida en el conjunto de prueba para calcular la pérdida promedio después de evaluar todo el conjunto de prueba.

Al final de la evaluación en el conjunto de prueba, se imprime la pérdida promedio en el conjunto de prueba y la precisión en el conjunto de prueba.



```python
# Evaluación de la red en el conjunto de prueba
model.eval()

test_loss = 0.0
correct = 0
total = 0

with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)

        # Calcular la salida y la pérdida
        outputs = model(images)
        loss = nn.CrossEntropyLoss()(outputs, labels)

        # Calcular la precisión en el conjunto de prueba
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

        test_loss += loss.item()

    # Imprimir métricas del conjunto de prueba
    print(f"Test Loss: {test_loss/len(test_loader):.4f}, Test Accuracy: {(correct/total)*100:.2f}%")
```

```python

import numpy as np
from sklearn.metrics import confusion_matrix
import seaborn as sns

predictions = []
true_labels = []

with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)

        predictions.extend(predicted.tolist())
        true_labels.extend(labels.tolist())

# Calcular la matriz de confusión
confusion_mat = confusion_matrix(true_labels, predictions)

# Visualizar la matriz de confusión
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_mat, annot=True, cmap="Blues", fmt="d")
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

## Visualización de resultados

```python
import matplotlib.pyplot as plt
import numpy as np

# Obtener una muestra del conjunto de prueba
data_iter = iter(test_loader)
images, labels = next(data_iter)
images = images.to(device)

# Realizar la predicción en la muestra
outputs = model(images)
_, predicted = torch.max(outputs, 1)

# Obtener la etiqueta correspondiente a la predicción
class_names = ['Avión', 'Automóvil', 'Pájaro', 'Gato', 'Ciervo', 'Perro', 'Rana', 'Caballo', 'Barco', 'Camión']
predicted_labels = [class_names[pred] for pred in predicted]

# Mostrar la imagen y la predicción
fig, axes = plt.subplots(figsize=(10, 4), ncols=4)
for i, ax in enumerate(axes):
    image = images[i] / 2 + 0.5  # Desnormalizar la imagen
    np_image = image.cpu().numpy()  # Convertir la imagen a un arreglo NumPy
    np_image = np.transpose(np_image, (1, 2, 0))  # Cambiar el orden de los canales (de C x H x W a H x W x C)
    ax.imshow(np_image)
    ax.set_title(predicted_labels[i])
    ax.axis('off')

plt.tight_layout()
plt.show()
```
